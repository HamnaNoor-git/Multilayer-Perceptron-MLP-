```python
import numpy as np

def sigmoid(z):
    return 1/(1+np.exp(-z))


# weights
w1, w2, w3, w4 = 0.67, 0.60, 0.89, 0.60
w5, w6, w7, w8 = -0.77, -0.88, -0.21, -0.63
w9, w10 = 0.72, -0.10

# biases
b1 = -0.43
b2 = -0.35


def forward(x1, x2):

    z1_1 = w1*x1 + w3*x2 + b1
    a1_1 = sigmoid(z1_1)

    z2_1 = w2*x1 + w4*x2 + b1
    a2_1 = sigmoid(z2_1)

    z1_2 = w5*a1_1 + w7*a2_1 + b2
    a1_2 = sigmoid(z1_2)

    z2_2 = w6*a1_1 + w8*a2_1 + b2
    a2_2 = sigmoid(z2_2)

    z1_3 = w9*a1_2 + w10*a2_2
    y_hat = sigmoid(z1_3)

    return y_hat


# Outside function
samples = [(0.2, 0.8), (0.9, 0.4), (0.1, 0.2)]

for i, (x1, x2) in enumerate(samples):
    print(f"Sample {i+1}: {forward(x1, x2):.4f}")
```
